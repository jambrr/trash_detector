```python
import os
import shutil
import numpy as np
import tqdm
from ultralytics import YOLO
```

```python
from pycocotools.coco import COCO
data_source = COCO(annotation_file='./datasets/TACO/new_ann.json')
```

```python
label_transfer = {5: 0, 12: 1}
```

```python


img_ids = data_source.getImgIds()

catIds = data_source.getCatIds()
categories = data_source.loadCats(catIds)
categories.sort(key=lambda x: x['id'])
classes = {}
coco_labels = {}
coco_labels_inverse = {}
for c in categories:
    coco_labels[len(classes)] = c['id']
    coco_labels_inverse[c['id']] = len(classes)
    classes[c['name']] = len(classes)

class_num = {}

print(classes)
```

```python
#!mkdir -p ./datasets/tmp/labels ./datasets/tmp/images
save_base_path  = './datasets/trashnet/tmp/labels/'
save_image_path = './datasets/trashnet/tmp/images/'
```

```python
os.makedirs(save_base_path, exist_ok=True)

for index, img_id in tqdm.tqdm(enumerate(img_ids), desc='change .json file to .txt file'):
    img_info = data_source.loadImgs(img_id)[0]
    # å°‡å«æ–‡ä»¶å¤¾çš„è·¯å¾‘ä¿®æ”¹ç‚ºæ–‡ä»¶å
    save_name = img_info['file_name'].replace('/', '_')
    # ç§»å»æ–‡ä»¶æ“´å±•å
    file_name = save_name.split('.')[0]
    # ç²å–å–®å¼µåœ–åƒçš„å¯¬å’Œé«˜
    height = img_info['height']
    width = img_info['width']
    # è½‰æ›æ‰€å¾—txtæ–‡ä»¶å­˜å„²è·¯å¾‘
    save_path = save_base_path + file_name + '.txt'
    is_exist = False  # è¨˜éŒ„åœ–ç‰‡æ˜¯å¦åŒ…å«ç›®æ¨™åƒåœ¾é¡å‹å°è±¡
    with open(save_path, mode='w') as fp:
        # æ ¹æ“šåœ–ç‰‡ç·¨è™Ÿæ‰¾å‡ºåƒåœ¾å°è±¡çš„ç·¨è™Ÿé›†åˆ
        annotation_id = data_source.getAnnIds(img_id)
        boxes = np.zeros((0, 5))
        if len(annotation_id) == 0:  # é›†åˆå¤§å°ç‚º0
            fp.write('')
            continue
        # ç²å–cocoæ ¼å¼çš„æ¨™ç±¤
        annotations = data_source.loadAnns(annotation_id)
        lines = ''  # è¨˜éŒ„è½‰æ›å¾Œyoloæ ¼å¼çš„æ¨™ç±¤
        # éæ­·å°è±¡æ¨™ç±¤é›†
        for annotation in annotations:
            is_exist = True
            box = annotation['bbox']
            if box[2] < 1 or box[3] < 1:
                # å¦‚æœåŸæ¨™ç±¤ä¸­å‡ºç¾ç„¡é•·æˆ–å¯¬æ•¸æ“šçš„æƒ…æ³å‰‡è·³é
                continue
            # top_x,top_y,width,height==>cen_x,cen_y,width,height
            box[0] = round((box[0] + box[2] / 2) / width, 6)
            box[1] = round((box[1] + box[3] / 2) / height, 6)
            box[2] = round(box[2] / width, 6)
            box[3] = round(box[3] / height, 6)
            if label not in class_num.keys():
                class_num[label] = 0
            class_num[label] += 1
            lines = lines + str(label)  # å…ˆå­˜å„²æ¨™ç±¤
            for i in box:  # å†å­˜å„²ä½ç½®ä¿¡æ¯
                lines += ' ' + str(i)
            lines += '\n'  # æ›è¡Œ
        fp.writelines(lines)
    if is_exist:
        # å­˜åœ¨ç›®æ¨™é¡å‹å°è±¡ï¼Œå‰‡æ‹·è²åœ–åƒè‡³æŒ‡å®šç›®éŒ„
        shutil.copy('./datasets/trashnet/train/images/{}'.format(img_info['file_name']), os.path.join(save_image_path, save_name))
    else:
        # ä¸å­˜åœ¨å‰‡åˆªé™¤æ‰€ç”Ÿæˆçš„æ¨™ç±¤æ–‡ä»¶
        os.remove(save_path)
```
