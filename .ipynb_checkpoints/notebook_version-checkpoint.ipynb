```python
from ultralytics import YOLO
from ultralytics.data.converter import convert_coco
import json
import os
import re
```

```python
from pycocotools.coco import COCO
data_source = COCO(annotation_file='./datasets/TACO/annotations.json')

img_ids = data_source.getImgIds()

catIds = data_source.getCatIds()
categories = data_source.loadCats(catIds)
categories.sort(key=lambda x: x['id'])
classes = {}
coco_labels = {}
coco_labels_inverse = {}
for c in categories:
    coco_labels[len(classes)] = c['id']
    coco_labels_inverse[c['id']] = len(classes)
    classes[c['name']] = len(classes)

class_num = {}
print(classes)
```

```python
import os
from PIL import Image
import matplotlib.pyplot as plt

image_dir = "./datasets/taconet/train/images"
max_images = 100
image_files = [f for f in os.listdir(image_dir) if f.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.gif'))]

if not image_files:
    print("No images found in the directory.")

# Display each image
for i, image_file in enumerate(image_files[:max_images]):
    img_path = os.path.join(image_dir, image_file)
    try:
        # Open and display the image
        img = Image.open(img_path)
        plt.figure(figsize=(6, 6))
        plt.imshow(img)
        plt.axis('off')
        plt.title(image_file)
        plt.show()
    except Exception as e:
        print(f"Error displaying image {img_path}: {e}")
```
